```python
# Cut events in old but not in new
nCut2,nSwap2 = 0,0
chListCut2 = []
runListCut2 = []
energyListCut2 = []
nDC, nMu, nLN, nAvsE, nDCR = 0, 0, 0, 0, 0
for index, row in dfOldCut.iterrows():
    cut = 'run==%d and channel==%d'%(row['run'],row['channel'])
    if row['Energy'] < 500:
        continue
    if not dfNewCut.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-2,row['Energy']+2)).empty:
        continue
    elif not dfNewCut.query(cut+'and %.3f<OEnergy<%.3f'%(row['OEnergy']-5,row['OEnergy']+5)).empty:
        continue
    elif row['channel'] == 594:
        continue
    elif not dfNew.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).empty:
        dfTemp = dfNew.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).drop(['DateSec','Date','ClockTime','OEnergy','LocalTime'], axis=1)
        print dfTemp
        if dfTemp['isLNFill2'].values != 0 or dfTemp['isLNFill1'].values != 0:
            nLN += 1
        elif dfTemp['muVeto'].values != 0:
            nMu += 1
        elif dfTemp['wfDCBits'].values != 0:
            nDC += 1
        elif dfTemp['AvsE'].values <= -1:
            nAvsE += 1
        elif dfTemp['DCR'].values >= 0.:
            nDCR += 1
        print 'AvsE:%.3f  DCR:%.5f  Energy:%.3f  channel:%d  run:%d'%(row['AvsE'], row['DCR'], row['Energy'], row['channel'], row['run'])
        nSwap2 += 1
        chListCut2.append(row['channel'])
        runListCut2.append(row['run'])
        energyListCut2.append(row['Energy'])        
        continue
    elif row['run'] in dfNewCut['run'].values and row['channel'] in dfNewCut['channel'].values:
#     else:
        print 'Untagged - run: %d  channel: %d  Energy: %.3f  AvsE: %.3f  DCR: %.5f'%(row['run'], row['channel'], row['Energy'], row['AvsE'], row['DCR'])

        nCut2 += 1
print "Cut by LNFill: %d   muVeto: %d   wfDCBits: %d   AvsE: %d   DCR: %d"%(nLN, nMu, nDC, nAvsE, nDCR)
print "Total swapped: ", nSwap2
print "Total mismatch: ", nCut2
```

```python
# Cut events not rejected in new skim
nCut,nSwap = 0,0
chListCut = []
chListCutAvsE = []
chListCutDCR = []
chListCutBurst = []
runListCut = []
runListCutAvsE = []
runListCutDCR = []
runListCutBurst = []
energyListCut = []
energyListCutAvsE = []
energyListCutDCR = []
energyListCutBurst = []
nDC, nMu, nLN, nAvsE, nDCR, nBurst = 0, 0, 0, 0, 0, 0
for index, row in dfNewCut.iterrows():
    cut = 'run==%d and channel==%d'%(row['run'],row['channel'])
    if row['Energy'] < 500 or row['Energy'] > 2000:
        continue
    if not dfOldCut.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-2,row['Energy']+2)).empty:
        continue
#     elif not dfOldCut.query(cut+'and %.3f<OEnergy<%.3f'%(row['OEnergy']-5,row['OEnergy']+5)).empty:
#         continue
    elif row['channel'] == 594:
        continue
    elif not dfOld.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-2,row['Energy']+2)).empty:
        dfTemp = dfOld.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-2,row['Energy']+2))
        print dfTemp.drop(['DateSec','Date','ClockTime','OEnergy'], axis=1)
        if 7370e3<dfTemp['ClockTime'].values<7371e3 or 2192e3<dfTemp['ClockTime'].values<2195e3 or 7840e3<dfTemp['ClockTime'].values<7860e3:
            nBurst += 1
        elif 8384e3<dfTemp['ClockTime'].values<8387e3 or 8984e3<dfTemp['ClockTime'].values<8985e3 or 9002e3<dfTemp['ClockTime'].values<9005e3:
            nBurst += 1
            energyListCutBurst.append(row['Energy'])
            runListCutBurst.append(row['run'])
            chListCutBurst.append(row['channel'])
        elif dfTemp['isLNFill1'].values != 0:
            nLN += 1
        elif dfTemp['muVeto'].values != 0:
            nMu += 1
        elif dfTemp['wfDCBits'].values != 0:
            nDC += 1
        elif dfTemp['AvsE'].values <= 1:
            energyListCutAvsE.append(row['Energy'])
            runListCutAvsE.append(row['run'])
            chListCutAvsE.append(row['channel'])
            print 'AvsE Swapped -- AvsE:%.3f  DCR:%.5f  Energy:%.3f  channel:%d  run:%d'%(row['AvsE'], row['DCR'], row['Energy'], row['channel'], row['run'])
            nAvsE += 1
        elif dfTemp['DCR'].values >= 0.:
            energyListCutDCR.append(row['Energy'])
            runListCutDCR.append(row['run'])
            chListCutDCR.append(row['channel'])
            nDCR += 1
            print 'DCR Swapped -- AvsE:%.3f  DCR:%.5f  Energy:%.3f  channel:%d  run:%d'%(row['AvsE'], row['DCR'], row['Energy'], row['channel'], row['run'])
            
        chListCut.append(row['channel'])
        runListCut.append(row['run'])
        energyListCut.append(row['Energy'])
        nSwap += 1
        continue
    elif row['run'] in dfOldCut['run'].values and row['channel'] in dfOldCut['channel'].values:
#     else:
        print 'Untagged - run: %d  channel: %d  Energy: %.3f  AvsE: %.3f  DCR: %.5f LocalTime: %.2f'%(row['run'], row['channel'], row['Energy'], row['AvsE'], row['DCR'], row['LocalTime'])
        nCut += 1
print "Cut by LNFill: %d   muVeto: %d   wfDCBits: %d   AvsE: %d   DCR: %d  Burst: %d"%(nLN, nMu, nDC, nAvsE, nDCR, nBurst)
print "Total swapped: ", nSwap
print "Total mismatch: ", nCut
```
