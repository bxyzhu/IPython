```python
# Event list comparison
import numpy as np
import pandas as pd

# pd.options.display.max_rows = 999

import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load files
extDir = '/Users/brianzhu/macros/code/MJDAnalysis/ROI/'
dfNew = pd.read_csv(extDir+'NewList_DS3_full.csv', sep=',', index_col=0)
dfOld = pd.read_csv(extDir+'OldList_DS3_full.csv', sep=',', index_col=0)

dfNewCut = pd.read_csv(extDir+'NewList_DS3_allcuts.csv', sep=',', index_col=0)
dfOldCut = pd.read_csv(extDir+'OldList_DS3_allcuts.csv', sep=',', index_col=0)

dfNewFull = pd.read_csv(extDir+'NewList_DS3_nocuts.csv', sep=',', index_col=0)
dfOldFull = pd.read_csv(extDir+'OldList_DS3_nocuts.csv', sep=',', index_col=0)
```

```python
# In old but not in new
n = 0
chList = []
runList = []
energyList = []
for index, row in dfOld.iterrows():
    cut = 'run==%d and channel==%d'%(row['run'],row['channel'])
    if not dfNew.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).empty:
        continue
    elif not dfNew.query(cut+'and %.3f<OEnergy<%.3f'%(row['OEnergy']-5,row['OEnergy']+5)).empty:
        continue
    elif row['run'] in dfNew['run'].values and row['channel'] in dfNew['channel'].values: 
#     else:
        print 'run: %d  channel: %d  Energy: %.3f  AvsE: %.3f  DCR: %.5f'%(row['run'], row['channel'], row['Energy'], row['AvsE'], row['DCR'])
        chList.append(row['channel'])
        runList.append(row['run'])
        energyList.append(row['Energy'])
        n += 1
print "Total mismatched events: ", n
```

```python
# In new but not in old
n2 = 0
chList2 = []
runList2 = []
energyList2 = []
for index, row in dfNew.iterrows():
    cut = 'run==%d and channel==%d'%(row['run'],row['channel'])
    if not dfOld.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).empty:
        continue
    elif not dfOld.query(cut+'and %.3f<OEnergy<%.3f'%(row['OEnergy']-5,row['OEnergy']+5)).empty:
        continue
    elif row['run'] in dfOld['run'].values and row['channel'] in dfOld['channel'].values:
#     else:
        print 'run: %d  channel: %d  Energy: %.3f  AvsE: %.3f  DCR: %.5f LocalTime: %.2f'%(row['run'], row['channel'], row['Energy'], row['AvsE'], row['DCR'], row['LocalTime'])
        chList2.append(row['channel'])
        runList2.append(row['run'])
        energyList2.append(row['Energy'])
        n2 += 1
print "Total mismatch: ", n2
```

```python
# Since mismatched events, look in full file
# In new but not in old
nfull = 0
chListfull = []
runListfull = []
energyListfull = []
for run,channel,energy in zip(runList2, chList2, energyList2):
    cut = 'run==%d and channel==%d'%(run, channel)
    if not dfOldFull.query(cut+'and %.3f<Energy<%.3f'%(energy-5,energy+5)).empty:
        print 'Found events -- run: %d  channel: %d  Energy: %.3f'%(run, channel, energy)
        print dfOldFull.query(cut+'and %.3f<Energy<%.3f'%(energy-5,energy+5)).drop(['DateSec','Date','ClockTime','OEnergy'], axis=1)
        continue
    elif run in dfOldFull['run'].values and channel in dfOldFull['channel'].values:
#         print 'run: %d  channel: %d  Energy: %.3f  AvsE: %.3f  DCR: %.5f LocalTime: %.2f'%(row['run'], row['channel'], row['Energy'], row['AvsE'], row['DCR'], row['LocalTime'])
        print 'Mismatched events -- run: %d  channel: %d  Energy: %.3f'%(run, channel, energy)
        chListfull.append(row['channel'])
        runListfull.append(row['run'])
        energyListfull.append(row['Energy'])
        nfull += 1
print "Total mismatch: ", nfull
```

```python
# Cut events in old but not in new
nCut2,nSwap2 = 0,0
chListCut2 = []
runListCut2 = []
energyListCut2 = []
nDC, nMu, nLN, nAvsE, nDCR = 0, 0, 0, 0, 0
for index, row in dfOldCut.iterrows():
    cut = 'run==%d and channel==%d'%(row['run'],row['channel'])
    if row['Energy'] < 500:
        continue
    if not dfNewCut.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).empty:
        continue
    elif not dfNewCut.query(cut+'and %.3f<OEnergy<%.3f'%(row['OEnergy']-5,row['OEnergy']+5)).empty:
        continue
    elif not dfNew.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).empty:
        dfTemp = dfNew.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).drop(['DateSec','Date','ClockTime','OEnergy','LocalTime'], axis=1)
        print dfTemp
        if dfTemp['isLNFill2'].values == 1 or dfTemp['isLNFill1'].values == 1:
            nLN += 1
        elif dfTemp['muVeto'].values != 0:
            nMu += 1
        elif dfTemp['wfDCBits'].values != 0:
            nDC += 1
        elif dfTemp['AvsE'].values <= -1:
            nAvsE += 1
        elif dfTemp['DCR'].values >= 0.:
            nDCR += 1
        print 'AvsE:%.3f  DCR:%.5f  Energy:%.3f  channel:%d  run:%d'%(row['AvsE'], row['DCR'], row['Energy'], row['channel'], row['run'])
        nSwap2 += 1
        chListCut2.append(row['channel'])
        runListCut2.append(row['run'])
        energyListCut2.append(row['Energy'])        
        continue
    elif row['run'] in dfNewCut['run'].values and row['channel'] in dfNewCut['channel'].values:
#     else:
        print 'Untagged - run: %d  channel: %d  Energy: %.3f  AvsE: %.3f  DCR: %.5f LocalTime: %.2f'%(row['run'], row['channel'], row['Energy'], row['AvsE'], row['DCR'], row['LocalTime'])

        nCut2 += 1
print "Cut by LNFill: %d   muVeto: %d   wfDCBits: %d   AvsE: %d   DCR: %d"%(nLN, nMu, nDC, nAvsE, nDCR)        
print "Total swapped: ", nSwap2
print "Total mismatch: ", nCut2
```

```python
# Cut events in new but not in old
nCut,nSwap = 0,0
nMu, nAvsE, nDCR = 0, 0, 0
chListCut = []
runListCut = []
energyListCut = []
nDC, nMu, nLN, nAvsE, nDCR = 0, 0, 0, 0, 0
for index, row in dfNewCut.iterrows():
    cut = 'run==%d and channel==%d'%(row['run'],row['channel'])
    if row['Energy'] < 500:
        continue
    if not dfOldCut.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).empty:
        continue
    elif not dfOldCut.query(cut+'and %.3f<OEnergy<%.3f'%(row['OEnergy']-5,row['OEnergy']+5)).empty:
        continue
    elif not dfOld.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).empty:
        dfTemp = dfOld.query(cut+'and %.3f<Energy<%.3f'%(row['Energy']-5,row['Energy']+5)).drop(['DateSec','Date','ClockTime','OEnergy'], axis=1)
        print dfTemp
        print 'AvsE:%.3f  DCR:%.5f  Energy:%.3f  channel:%d  run:%d'%(row['AvsE'], row['DCR'], row['Energy'], row['channel'], row['run'])
        if dfTemp['isLNFill2'].values == 1 or dfTemp['isLNFill1'].values == 1:
            nLN += 1
        elif dfTemp['muVeto'].values != 0:
            nMu += 1
        elif dfTemp['wfDCBits'].values != 0:
            nDC += 1
        elif dfTemp['AvsE'].values <= -1:
            nAvsE += 1
        elif dfTemp['DCR'].values >= 0.:
            nDCR += 1
        chListCut.append(row['channel'])
        runListCut.append(row['run'])
        energyListCut.append(row['Energy'])
        nSwap += 1
        continue
    elif row['run'] in dfOldCut['run'].values and row['channel'] in dfOldCut['channel'].values:
#     else:
        print 'Untagged - run: %d  channel: %d  Energy: %.3f  AvsE: %.3f  DCR: %.5f LocalTime: %.2f'%(row['run'], row['channel'], row['Energy'], row['AvsE'], row['DCR'], row['LocalTime'])
        nCut += 1
print "Cut by LNFill: %d   muVeto: %d   wfDCBits: %d   AvsE: %d   DCR: %d"%(nLN, nMu, nDC, nAvsE, nDCR)
print "Total swapped: ", nSwap
print "Total mismatch: ", nCut
```
