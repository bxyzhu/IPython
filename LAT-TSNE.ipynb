```python
# Read in DS1 Force Trigger files
# Use Ch 626 (same as Ch 624 in DS0)
inDir = '/Users/brianzhu/project/Pandas/ForceTrigger'
bkgFiles = glob.glob('%s/latSkimDS1_run99*.h5' % inDir)
dfBkgFull = pd.DataFrame()
bkgList = []
for file in bkgFiles:
    print(file)
    dfDummy = pd.read_hdf(file,'skimTree')
    bkgList.append(dfDummy)

dfBkgFull = pd.concat(bkgList)
dfBkgFull.head()
```

```python
# Read in Data files from DS1
bkgDir = '/Users/brianzhu/project/Pandas/bkg'
dataFiles = glob.glob('%s/latSkimDS1_*.h5' % bkgDir)
dfDataFull = pd.DataFrame()
dataList = []
for file in dataFiles:
    print(file)
    dfDummy = pd.read_hdf(file,'skimTree')
    dataList.append(dfDummy)

dfDataFull = pd.concat(dataList)
dfDataFull.head()
```

```python
# Clean up datasets a little
# Determined that the cut to use to remove extra pulsers for FT run is oppie < 15 || TMath::IsNaN(oppie)
# First replace all NaN values in oppie with a dummy variable like -9999, this allows a cut like oppie < 15 to remove pulsers

# dfBkg.loc[:,'S5Cut'] = df1.loc[:,'waveS5'] / np.power(df1.loc[:,'trapENFCal'], 0.25)
# dfBkg = np.isnan(dfBkg.loc[:,'oppie'])

# inplace=True will replaces the values in the column with the new values
dfBkgFull['oppie'].fillna(-9999, inplace=True)
print (dfBkg.shape)

# Check for NaN -- those suck
print dfBkgFull.isnull().sum()
print dfSigFull.isnull().sum()
print dfDataFull.isnull().sum()
```

```python
# The background DF is set, the signal DF is already good to go
dfBkgTest = dfBkgFull.query('oppie < 15 and channel==626')
dfSig = dfSigFull.query('channel==624')
dfData = dfDataFull.query('channel==626')
dfBkgTest.shape
print ('Force Trigger: ', dfBkgTest.shape)
print ('Ext Pulser: ', dfSig.shape)
print ('Data: ', dfData.shape)
print (dfSig.columns.values)
```

```python
# Drop all columns that aren't useful for learning -- drop all time parameters since we don't trust them
dfBkgTest = dfBkgTest.drop(['C','channel','wpar4','bcMin','run','fails','bandTime','den10','den50','den90','derivTime','exp0','fitE','fitMatch','fitMax','kvorrT','pol0','raw10','raw50','raw90','threshKeV','threshSigma','trapENF','trapENM','trapENMCal','trapENMSample','triggerTrapt0','waveS0'], axis=1)
dfSig = dfSig.drop(['C','channel','wpar4','bcMin','run','fails','bandTime','den10','den50','den90','derivTime','exp0','fitE','fitMatch','fitMax','kvorrT','pol0','raw10','raw50','raw90','threshKeV','threshSigma','trapENF','trapENM','trapENMCal','trapENMSample','triggerTrapt0','waveS0'], axis=1)
dfData = dfData.drop(['C','channel','wpar4','bcMin','run','fails','bandTime','den10','den50','den90','derivTime','exp0','fitE','fitMatch','fitMax','kvorrT','pol0','raw10','raw50','raw90','threshKeV','threshSigma','trapENF','trapENM','trapENMCal','trapENMSample','triggerTrapt0','waveS0'], axis=1)

# Create array of parameter names to train on, remove trapENFCal (don't want to train on energy)
feat_cols = np.delete(dfBkgTest.columns.values, np.argwhere(dfBkgTest.columns.values=='trapENFCal' ))

# Add Column of zeros and ones for truth information
dfBkgTest['Truth'] = 'Force Trigger'
dfSig['Truth'] = 'Ext Pulser'
dfData['Truth'] = 'DS1 data'
print ("Columns: ", dfBkgTest.columns.values)
print ("Features to train on: ",feat_cols)
```

```python
# Split up training data for testing
from sklearn.model_selection import train_test_split

TotalDF = pd.concat([dfBkgTest,dfSig,dfData])
TotalDF.head()
print (TotalDF.shape)

# Split it twice, first split creates development (with 66% of data) and evaluation set
# 2nd split is for training and testing the training
DF_dev, DF_eval = train_test_split(TotalDF, test_size=0.33, random_state=42)
DF_train, DF_test = train_test_split(DF_dev, test_size=0.33, random_state=492)

print(DF_dev.shape, DF_eval.shape, DF_train.shape, DF_test.shape)
print(DF_dev.columns.values)

# For sklearn, drop the Truth column for X, only save Truth for Y
# X is the data, Y is the target
X_dev = DF_dev.drop(['Truth', 'trapENFCal'], axis=1)
X_eval = DF_eval.drop(['Truth', 'trapENFCal'], axis=1)
X_train = DF_train.drop(['Truth', 'trapENFCal'], axis=1)
X_test = DF_test.drop(['Truth', 'trapENFCal'], axis=1)
Y_dev = DF_dev['Truth']
Y_eval = DF_eval['Truth']
Y_train = DF_train['Truth']
Y_test = DF_test['Truth']
print(X_dev.shape, X_eval.shape, X_train.shape, X_test.shape)
print(Y_dev.shape, Y_eval.shape, Y_train.shape, Y_test.shape)
print(X_dev.columns.values)
```

```python
# Try PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=10)
pca_result = pca.fit_transform(DF_dev.loc[:,feat_cols].values)
# pca_result = pca.fit_transform(X_dev)

DF_dev.loc[:,'pca-one'] = pd.Series(pca_result[:,0])
DF_dev.loc[:,'pca-two'] = pd.Series(pca_result[:,1])
DF_dev.loc[:,'pca-three'] = pd.Series(pca_result[:,2])
DF_dev.loc[:,'pca-four'] = pd.Series(pca_result[:,3])
DF_dev.loc[:,'pca-five'] = pd.Series(pca_result[:,4])
DF_dev.loc[:,'pca-six'] = pd.Series(pca_result[:,5])
DF_dev.loc[:,'pca-seven'] = pd.Series(pca_result[:,6])
DF_dev.loc[:,'pca-eight'] = pd.Series(pca_result[:,7])
DF_dev.loc[:,'pca-nine'] = pd.Series(pca_result[:,8])
DF_dev.loc[:,'pca-ten'] = pd.Series(pca_result[:,9])

print 'Explained variation per principal component: ',(pca.explained_variance_ratio_)
DF_dev.head()
```

```python
# Try TSNE by itself
import time
from sklearn.manifold import TSNE

# Limit the number of events to process for now
n_sne = 15000

time_start = time.time()
tsne = TSNE(n_components=2, verbose=2, perplexity=50, n_iter=500)
tsne_results = tsne.fit_transform(DF_dev.loc[:,feat_cols].values[:n_sne])

print 't-SNE done! Time elapsed: %.3f seconds'%(time.time()-time_start)
```
